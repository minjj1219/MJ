```python
corona_df.columns = ["인덱스", "등록일시", "사망자", "확진자", "게시글번호", "기준일", "기준시간","수정일시", "누적 의심자", "누적확진률"]

corona_df
```

```python
corona_df_2 = corona_df.sort_values("등록일시")

corona_df_2
```

```python
corona_df_2["일일 확진자"] = (corona_df_2["확진자"] - corona_df_2["확진자"].shift()).fillna(0)

corona_df_2
```

```python
# fillna(0) 결측치를 0으로 채워넣겠다

corona_df_2["일일 사망자"] = corona_df_2["사망자"].diff().fillna(0)

corona_df_2
```

```python
corona_df_2["일일 사망자"].value_counts(ascending=True)
```

```python
corona_df_3 = corona_df_2.drop(['인덱스', '기준일', '게시글번호', '기준시간', '수정일시'], axis=1)

corona_df_3
```

```python
corona_df_3.reset_index(drop=True, inplace=True)

corona_df_3
```

```python
corona_df_3
```

```python
corona_df_4 = corona_df_3.tail(50)
```

```python
corona_df_4.fillna(method="ffill")
# forward fill 앞에 값으로 채우기
```

```python
corona_df_4.fillna(method="bfill")
# back fill 뒤에 값으로 채우기
```
