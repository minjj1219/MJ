```python
import pandas as pd
```

```python
sales_df = pd.read_csv("./csv/Sales Records.csv")
sales_df.head()
```

```python
sales_df_2 = sales_df.drop(['Ship Date','Order ID','Sales Channel','Total Revenue'],axis=1)

sales_df_2
```

```python
sales_df_2.columns=['지역','국가','상품타입','배송순위','주문날짜','판매개수','단가','원가','총비용','총이윤']
sales_df_2
```

```python
sales_df_2.sort_values("국가", inplace=True)
sales_df_2
```

```python
sales_df_2.reset_index(drop=True, inplace=True)
sales_df_2
```

```python
#정수형으로 변경
sales_df_2["총이윤"] = sales_df_2["총이윤"].astype(int)
```

```python
#반올림 정수화/ astype(int) 안넣으면 뒤에 .0붙음
#round 안의 숫자= 소수점의 개수
sales_df_2["총비용"].round(0).astype(int)
```

```python
sales_df["원가"] = sales_df_2["원가"].round(1)
```

```python
sales_df_2["판매개수"].round(-1)
```

```python
#format은 원본 데이터와 맞춰줘야함(연월일 순서, 기호)
sales_df_2["주문날짜"] = pd.to_datetime(sales_df_2["주문날짜"],
                         format="%m/%d/%Y")
sales_df_2.head()
```

```python
#지역에 해당하는 국가별로 나눠서 총이윤 합 구함
sales_df_2.groupby(["지역","국가"]).sum()["총이윤"]
```

```python
sales_df_2.groupby("국가").get_group("South Korea")

```

```python
sales_df_2.groupby(['국가'])['총이윤'].mean()[['South Korea','China','Togo']].mean()
```

```python
sales_df_3 = sales_df_2[sales_df_2['국가'].isin(["South Korea", "China", "Togo"])]
sales_df_3.groupby("국가").mean()["총이윤"]

```
