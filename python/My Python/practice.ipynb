```python
def func_11(a,b):
    result_1 = a+b
    return result_1

def func_12(a,b):
    result_2 = a-b
    return result_2

def func_13(a,b,c):
    if c == 'S':
        print(func_11(a,b))
    elif c == 'D':
        print(func_12(a,b))

func_13(1,2,'S')

# print(func_13(1,2,'S')) -> 함수 안에 print 있으면 none도 함께 출력됨
# none 안나오게 하려면 함수 안에 print 말고 return 쓰기
# return 없으면 반환값이 없기 때문에 none이 출력되는 것임
```

```python
import pandas as pd

corona_df = pd.read_csv('corona.csv')
corona_df
```

```python
corona_df.columns = ['인덱스', '등록일시', '사망자', '확진자', '게시글번호', '기준일', '기준시간', '수정일시', '누적 의심자', '누적확진률']
corona_df
```

```python
corona_df.describe()
```

```python
corona_df.사망자
corona_df['확진자']
```

```python
corona_df_2 = corona_df.sort_values('등록일시')

corona_df_2
```

```python
corona_df_2['일일 확진자'] = (corona_df_2['확진자']-corona_df_2["확진자"].shift()).fillna(0)

corona_df_2
```

```python
corona_df_3 = corona_df.copy().sort_values("등록일시")

corona_df_3['일일 사망자'] = corona_df_2['사망자'].diff()

corona_df_3

```

```python
corona_df_2['일일 확진자'].value_counts()
# value_counts()는 컬럼 값의 분포도를 나타냄
```

```python
corona_df_4 = corona_df_2.drop(['인덱스', '기준시간','누적 의심자','누적확진률', '게시글번호','수정일시'], axis=1)

corona_df_4
```

```python
corona_df_4.reset_index(drop=True)
#인덱스 초기화하여 수정된 df에 인덱스 0부터 다시 부여
```
