```python
#for i in range(1,5,1):
    
google_1= google.round(2)
google_1
```

```python
google_1.isna().sum()
```

```python
google_1["등락폭"] = google_1["Close"]-google_1["Open"]
google_1["등락율"] = google_1["등락폭"]/google_1["Open"]*100
google_1
```
