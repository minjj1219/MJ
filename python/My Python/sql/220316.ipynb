```python
# CASE는 여러값 동시에 바꿔줌(조건 여러개 사용 가능)
# ELSE도 사용가능
sql_16 = """
             SELECT ENAME,
             CASE
                 WHEN JOB = "ANALYST" THEN "분석가"
                 WHEN JOB = "CLERK" THEN "점원"
                 WHEN JOB = "MANAGER" THEN "관리자"
                 WHEN JOB = "SALESMAN" THEN "판매사원"
                 WHEN JOB = "PRESIDENT" THEN "사장"
                 END AS JOB
             FROM emp          
        """

result_16 = _dt.executeAll(sql_16)
df_16 = pd.DataFrame(result_16)
df_16
```

1. emp table과 dept table을 left join 조건 dept가 같은 경우
2. 출력 컬럼은 ENAME, JOB, SAL, DEPTNO, LOC
3. ENAME, LOC 그대로 출력
4. JOB은 각각 한글명으로 변경
5. SAL은 2450을 기준으로 2450 이상이면 HIGH 미만이면 LOW
6. DEPTNO은 DEPTNO의 값과 dept table의 DNAME 값을 더해서 표시 (DEPTNO-DNAME형식)
7. 모든 컬럼의 이름은 한글로 변경

```python
# 중복되는 컬럼 있으면 꼭 테이블명 앞에 적어줘야함!

sql_17 = """
            SELECT 
            emp.ENAME AS "사원명",
            CASE
                 WHEN emp.JOB = "ANALYST" THEN "분석가"
                 WHEN emp.JOB = "CLERK" THEN "점원"
                 WHEN emp.JOB = "MANAGER" THEN "관리자"
                 WHEN emp.JOB = "SALESMAN" THEN "판매사원"
                 WHEN emp.JOB = "PRESIDENT" THEN "사장"
            END AS "직급",
            IF(emp.SAL<2450, 'LOW', 'HIGH') AS "월급",
            CONCAT(dept.DEPTNO,'-',dept.DNAME) AS "부서번호-부서명",
            dept.LOC AS "부서명"
            FROM emp
            LEFT JOIN dept
            ON emp.DEPTNO = dept.DEPTNO
            ORDER BY 사원명 ASC
     
         """

result_17 = _dt.executeAll(sql_17)
df_17 = pd.DataFrame(result_17)
df_17
```

부서의 이름이 'sales'인 소속 사원들의 리스트를 출력하라

```python
sql_20 = """
            SELECT 
            ENAME 
            FROM emp
            LEFT JOIN dept
            ON emp.DEPTNO = dept.DEPTNO
            WHERE dept.DNAME = 'sales'
         """

result_20 = _dt.executeAll(sql_20)
pd.DataFrame(result_20)
```

```python
# 다중쿼리문
sql_21 = """
            SELECT
            ENAME
            FROM emp
            WHERE DEPTNO = 
            (
                SELECT LOC
                FROM dept
                WHERE DNAME = "sales"
            )

         """

result_21 = _dt.executeAll(sql_21)
pd.DataFrame(result_21)
```

### 부서의 위치가 'NEWYORK','DALLAS'가 아닌 소속 사원들의 리스트를 출력하라

```python
# 1242 에러 : = 조건에 충족하는 값이 두 개 이상일 때 나는 에러
# 이럴 때는 IN을 써야함
sql_22 = """
            SELECT
            ENAME
            FROM emp
            WHERE 
            DEPTNO IN
            (
                SELECT DEPTNO
                FROM dept
                WHERE LOC NOT IN ('NEW YORK','DALLAS')
            )

         """

result_22 = _dt.executeAll(sql_22)
pd.DataFrame(result_22)
```
